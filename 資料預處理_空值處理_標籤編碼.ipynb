### 使用 pd.to_datetime() 轉換 datetime 格式

```python
d3 = pd.DataFrame(df['交易年月日'].apply(lambda x: str(x+19110000)),columns=['交易年月日'])

# 替換 無交易月日值 為 一月一日
d4 = d3['交易年月日'].str.replace('00..$|0{4}$','0101')

# 驗證沒有無交易月日資料
md04 = d4.str.contains('0{4}$').sum()
d02 = d4.str.contains('^00..$').sum()

print(f'取代後月日為0000有幾個 : {md04}')
print(f'取代後月份為00有幾個 : {d02}')
```

```python
# 轉成 datetime 格式
pd.to_datetime(d4)
```

```python
df['交易年月日'] = pd.to_datetime(d4)
df.head()
```

# 交易標的 轉 數值標籤

```python
transaction_subjects_mapping={'房地(土地+建物)':0,'房地(土地+建物)+車位':1,'建物':2}
df['交易標的']=df['交易標的'].map(transaction_subjects_mapping)
```

# 土地移轉總面積平方公尺 轉 土地移轉總面積坪

```python
df['土地移轉總面積坪'] = df['土地移轉總面積平方公尺'] / 0.3025 
```

# 建物移轉總面積平方公尺 轉 建物移轉總面積坪

```python
df['建物移轉總面積坪'] = df['建物移轉總面積平方公尺'] / 0.3025 
```

```python
df['車位移轉總面積(坪)'] = df['車位移轉總面積(平方公尺)'] / 0.3025
```

# 建物現況格局-隔間 轉 二元變數

```python
layout_mapping = {'無':0, '有':1}
df['建物現況格局-隔間'] = df['建物現況格局-隔間'].map(layout_mapping)
df['建物現況格局-隔間']
```

# 有無管理組織 轉 二元變數

```python
manage_mapping = {'無': 0,'有': 1,}

df['有無管理組織'] = df['有無管理組織'].map(manage_mapping)
```

# 鄉鎮市區 轉 數值編碼

```python
list(enumerate(np.unique(df['鄉鎮市區'])))
```

```python
# 建立對應字典
# 將類標籤從字符串轉換為整數
class_mapping = {label: idx for idx, label in enumerate(np.unique(df['鄉鎮市區']))}
df['鄉鎮市區'] = df['鄉鎮市區'].map(class_mapping)
df['鄉鎮市區']
```

# 電梯 轉 二元變數

```python
elevator_mapping = {'無': 0,'有': 1}

df['電梯'] = df['電梯'].map(elevator_mapping)
```

# 縣市 轉 二元變數

```python
size_mapping = {'新北市': 0,'臺北市': 1,}

df['縣市'] = df['縣市'].map(size_mapping)
```
