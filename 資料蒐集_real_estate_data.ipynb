# 資料預處理

## 從本機路徑中讀取不動產實價登錄資料

```python
# 雙北
city =['a','f']

# 歷年各季資料夾
dirs = [dir for dir in os.listdir() if dir[:5] == 'real_']

dfs = []

for d in dirs:
    print(d)
    for c in city:
      df = pd.read_csv(os.path.join(d, c+"_lvr_land_a.csv"), index_col=False)
      df['Q'] = d[-1]
      dfs.append(df)

df = pd.concat(dfs, sort=True)
```

```python
# 把"建築完成年"切分出來放進list
d = df['建築完成年月']
complete_rocyear = []
for i in range(len(df)):
    if len(d.values[i]) == 3:  # 長度為3, 表示只有民國年
        complete_rocyear.append( d.values[i].strip() )
    elif len(d.values[i]) == 4:  # 長度為4, 例如5602
        complete_rocyear.append( (d.values[i].strip())[0:2] )  # 只取前兩個字元
    elif len(d.values[i]) == 5:
        complete_rocyear.append( (d.values[i].strip())[0:3] )
    elif len(d.values[i]) == 6:
        complete_rocyear.append( (d.values[i].strip())[0:2] )
    else:
        complete_rocyear.append( (d.values[i].strip())[0:3] )
```
