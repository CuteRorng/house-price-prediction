# 各行政區時間序列

```python
for i in range(41):
  if ((i != 13) & (i != 17) & (i != 30)):
    data = pd.read_csv('/content/drive/MyDrive/極端值處理及常態分佈圖/ML data_outliers deletion_IQR.csv')
    data['transaction date_yr'] = data['transaction date'].str[:7]
    data = data[['transaction date_yr','unit price_ping','district']]
    data = data.loc[data['district'] == i]
    t = data.drop(['district'], axis=1)
    t.rename(columns={'transaction date_yr': 'ds', 'unit price_ping': 'y'}, inplace=True)
    y = t.groupby('ds').mean()[['y']]
    y['ds'] = y.index
    y = y.reset_index(drop=True)
    y = y[['ds','y']]
    y = y[12:]
    y = y.reset_index(drop=True)

    prophet = Prophet(seasonality_mode='multiplicative', mcmc_samples=300)
    prophet.fit(y)

    # create a future data frame 
    # 構建待預測日期數據框，periods = 365 代表除歷史數據的日期外再往後推 365 天
    future = prophet.make_future_dataframe(periods=36, freq='MS')
    forecast = prophet.predict(future)

    # display the most critical output columns from the forecast
    forecast[['ds','yhat','yhat_lower','yhat_upper']].head()

    # plot
    # fig = prophet.plot(forecast)
    fig = prophet.plot_components(forecast)

    path = '/content/drive/MyDrive/ML/圖片/時間序列/分區域'
    file_path = os.path.join(path, f"time_{i}.png")
    fig.savefig(file_path ,bbox_inches = 'tight',dpi=700)

    ax = fig.gca()
    # ax.set_title("section", size=25)
    # ax.set_xlabel("Year", size=12)
    # ax.set_ylabel("Price/Ping", size=12)
    ax.tick_params(axis="x", labelsize=8)
    ax.tick_params(axis="y", labelsize=8)
    fig.savefig("section", dpi=200, bbox_inches='tight')

    # interactive plot 
    from prophet.plot import plot_plotly, plot_components_plotly
    import plotly.express as px

    fig1 = plot_plotly(prophet, forecast)  
    file_path = os.path.join(path, f"time_{i}.html")
    fig1.write_html(file_path)
```

# 大台北地區時間序列

```python
data = pd.read_csv('/content/drive/MyDrive/極端值處理及常態分佈圖/ML data_outliers deletion_IQR.csv')
data['transaction date_yr'] = data['transaction date'].str[:7]
t = data[['transaction date_yr','unit price_ping']]
t.rename(columns={'transaction date_yr': 'ds', 'unit price_ping': 'y'}, inplace=True)
y = t.groupby('ds').mean()[['y']]
y['ds'] = y.index
y = y.reset_index(drop=True)
y = y[['ds','y']]
y = y[12:]
y = y.reset_index(drop=True)
y
```

```python
# instantiate the model and fit the timeseries
# seasonality_mode='multiplicative': 乘法季節性
prophet = Prophet(seasonality_mode='multiplicative', mcmc_samples=300)
# yearly_seasonality = True,
# weekly_seasonality = True,
# daily_seasonality = True,
#changepoint_prior_scale = 0.1

prophet.fit(y)

# create a future data frame 
# 構建待預測日期數據框，periods = 365 代表除歷史數據的日期外再往後推 365 天
future = prophet.make_future_dataframe(periods=36, freq='MS')
forecast = prophet.predict(future)

# display the most critical output columns from the forecast
forecast[['ds','yhat','yhat_lower','yhat_upper']].head()

# plot
# fig = prophet.plot(forecast)
fig = prophet.plot_components(forecast)
fig.savefig("/content/drive/MyDrive/ML/圖片/時間序列/time_4.png",bbox_inches = 'tight',dpi=700)

ax = fig.gca()
# ax.set_title("section", size=25)
# ax.set_xlabel("Year", size=12)
# ax.set_ylabel("Price/Ping", size=12)
ax.tick_params(axis="x", labelsize=8)
ax.tick_params(axis="y", labelsize=8)
fig.savefig("section", dpi=200, bbox_inches='tight')

# interactive plot 
from prophet.plot import plot_plotly, plot_components_plotly
import plotly.express as px

fig1 = plot_plotly(prophet, forecast)
fig1.write_html("/content/drive/MyDrive/ML/圖片/時間序列/time_4.html")
```
